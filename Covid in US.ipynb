# Dataset Description

The dataset contains the situation of the COVID-19 Pandemic in the different states of the US as on 21st May 2020

# Features Description

```python
df = pd.read_csv("COVID in US.csv")
df.head()
```

Clearly Country_Region column does not contain any useful info for analysis so let's drop it

```python
plt.figure(figsize=(20,6))
sns.barplot(x="Province_State", y="Confirmed",data=df)
plt.xticks(rotation=90)
plt.title('Province/State wise confirmed cases of COVID-19')
```

The  number of confirmed cases is highest in New York followed by New Jeresy

```python
plt.figure(figsize=(24,6))
sns.barplot(x="Province_State", y="Deaths",data=df)
plt.xticks(rotation=90)
plt.title('Province/Statewise deaths due to COVID-19')
```

The  number of deaths due to COVID-19 is highest in New York followed by New Jeresy and then Massachusetts, Michigan in order

```python
plt.figure(figsize=(24,6))
sns.barplot(x="Province_State", y="Recovered",data=df)
plt.xticks(rotation=90)
plt.title('Province/State wise Recovered from COVID-19')
```

Recovery rate is highest in New York followed by Texas, Michigan, Louisiana, New Jeresy in order

```python
plt.figure(figsize=(24,6))
sns.barplot(x="Province_State", y="Active",data=df)
plt.xticks(rotation=90)
plt.title('Province/State wise Active cases of COVID-19')
```

Number of active cases is highest in new York followed by New Jeresy, Issinois in order

```python
plt.figure(figsize=(24,6))
sns.barplot(x="Province_State", y="Incident_Rate",data=df)
plt.xticks(rotation=90)
plt.title('Province/State wise Incident_Rate of COVID-19')
```

Incident_Rate (confirmed cases per 100,000 persons) is highest in New York followed by New Jeresy, Rhode Island, Massachusetts, Connecticut, District of Columbia

```python
plt.figure(figsize=(24,6))
sns.barplot(x="Province_State", y="People_Tested",data=df)
plt.xticks(rotation=90)
plt.title('Province/State wise People_Tested for COVID-19')
```

Number of People tested was highest in New York followed by California, Texas, Florida, Illinois, New Jeresy in order 

```python
plt.figure(figsize=(24,6))
sns.barplot(x="Province_State", y="People_Hospitalized",data=df)
plt.xticks(rotation=90)
plt.title('Province/State wise People_Hospitalized due to COVID-19')
```

Number of People hospitalized was highest in New York followed by Connecticut, Massachusetts, Florida in order

```python
plt.figure(figsize=(24,6))
sns.barplot(x="Province_State", y="Mortality_Rate",data=df)
plt.xticks(rotation=90)
plt.title('Province/State wise Mortality_Rate of COVID-19')
```

Mortality rate is very high in Northern Mariana Islands followed by Michigan, Connecticut, Virgin Islands, New York in order

```python
plt.figure(figsize=(24,6))
sns.barplot(x="Province_State", y="Testing_Rate",data=df)
plt.xticks(rotation=90)
plt.title('Province/State wise Testing_Rate for COVID-19')
```

Testing rate is highest in Rhode island followed by New York, North Dakota, Masachusetts in order

```python
plt.figure(figsize=(24,6))
sns.barplot(x="Province_State", y="Hospitalization_Rate",data=df)
plt.xticks(rotation=90)
plt.title('Province/State wise Hospitalization_Rate')
```

Hospitalization rate is highst in Conecticut followed by Kentucky, New York, Oregon in Order
